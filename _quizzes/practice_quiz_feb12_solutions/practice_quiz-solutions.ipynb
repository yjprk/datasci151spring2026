**Q3** Plotting inside a for loop

<font size = "3">

Using a `for` loop, generate 3 scatter plots using the car data from the file "features.csv" in the "quiz_data" subfolder. The scatter plots should plot:

- displacement vs. mpg
- weight vs. mpg
- acceleration vs. mpg

````python
# your code here

df_cars = pd.read_csv("quiz_data/features.csv")

y_var = 'mpg' # y-axis variable always the same

plot_variables = ['displacement', 'weight', 'acceleration']

for x_var in plot_variables:
    plt.scatter(x = df_cars[x_var], y = df_cars[y_var])
    plt.xlabel(x_var)
    plt.ylabel(y_var)
    plt.title(x_var + " vs. " + y_var)
    plt.show()
````

**Q6** Sum all positive elements

<font size = "3">

Below, we create a random `numpy` array with positive and negative numbers. Using a `for` loop with a `continue` statement, compute the sum of all the **positive** elements of the array

````python
N = 25
random_array = np.random.uniform(low = -5, high = 10, size = N)


###### Solution 1
pos_sum_1 = 0 

for val in random_array:
    if val <= 0:
        continue 
    pos_sum_1 = pos_sum_1 + val 

###### Solution 2
pos_sum_2 = 0

for i in range(N):
    if random_array[i] <= 0:
        continue 
    pos_sum_2 = pos_sum_2 + random_array[i]

print(pos_sum_1)
print(pos_sum_2)
````
