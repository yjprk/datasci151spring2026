**(d)** Piecewise-defined function

<font size = "3">

Consider the piecewise-defined expression

$$
y = 
\begin{cases}
0 & \text{if } x \le 0 \\[6pt]
\dfrac{1}{2}x & \text{if } 0 < x \le \tfrac{1}{2} \\[10pt]
\dfrac{1}{2}(1 - x) & \text{if } \tfrac{1}{2} < x \le 1 \\[6pt]
0 & \text{if } 1 < x
\end{cases}
$$

In the code cell below, a random number `x` is generated uniformly
between $-1/2$ and $3/2$. Using `if`, `elif`, and/or `else` statements, write code that defines a variable `y` that is determined by the expression above. Then print it to the screen.

##### Solution 1

````python
import random 

x = random.uniform(-0.5, 1.5)

# SOLUTION
if x <= 0:
    y = 0
elif x <= 0.5:
    y = 0.5*x 
elif x <= 1:
    y = 0.5*(1 - x)
else:
    y = 0


print(y)
````

##### Solution 2

````python
import random 

x = random.uniform(-0.5, 1.5)

# SOLUTION
if x <= 0 or x > 1:
    y = 0
elif x <= 0.5:
    y = 0.5*x 
elif x <= 1:
    y = 0.5*(1 - x)

print(y)
````

##### Solution 3

Suppose a cardiology team wants to flag patients who are at-risk of hypertension complications. They use the following rule:

- A patient is at risk if their systolic blood pressure is **greater than or equal to** 140 **and** they are **65 or older**

In the code block below, 3 lists are created containing information for 10 patients:

- `patient_id_lst` contains unique identification numbers for the 10 patients.

- `systolic_bp_lst` contains systolic blood pressure values for the 10 patients.

- `age_lst` contains the age (in years) for the 10 patients.

The ordering of the lists are consistent, so that patient "A1023" is 72 years old with a systolic blood pressure of 138. Patient "A1044" is 60 years old with a systolic blood pressure of 142, etc.

Using a **for** loop, append the patient IDs of all "at-risk" patients (those who meet the criteria listed above) to the initially empty list `at_risk`.

Then **print** `at_risk` to the screen.

##### Solution 1

````python
patient_id_lst   = ["A1023", "A1044", "B2331", "B2440", "C3001", "C3120", 
    "D4105", "D4119", "E5202", "E5307"]
systolic_bp_lst  = [138, 142, 152,  171, 155,  119, 144, 140, 132, 139]
age_lst          = [72,   60,   68,  65,  82,   34,  55,  70,  67, 75]

at_risk = []

bp_cut_off = 140
age_cut_off = 65

# SOLUTION
N = len(patient_id_lst)

for k in range(N):
    bp = systolic_bp_lst[k]
    id_no = patient_id_lst[k]
    age = age_lst[k]

    if (age >= age_cut_off) and (bp >= bp_cut_off):
        at_risk.append(id_no)

print(at_risk)
````
