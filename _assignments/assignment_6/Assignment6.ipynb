(a) Define a function to convert temperatures 

<font size = "3">

The formula for converting temperature from degrees Fahrenheit to degrees Kelvin is 
$$\frac{5}{9}(F-32)+273.15$$
where $F$ = temperature in degrees Fahrenheit
1. Define a function ```f_to_k``` with a single argument ``F`` which converts the input temperature (in Fahrenheit) to the equivalent value in Kelvin.


2. Use your function to convert the following Fahrenheit temperatures to Kelvin and print the values: <br>
- 32 degrees Fahrenheit (freezing point of water)
- 212 degrees Fahrenheit (boiling point of water)
- 165 degrees Fahrenheit (minimal safe internal temperature of cooked chicken)

````python
# your code here
def f_to_k(F):
    return (5/9) * (F - 32) + 273.15

print(f_to_k(32))   
print(f_to_k(212))  
print(f_to_k(165)) 
````

(b) Define a function to compute the sum of squares

<font size = "3">

In Quiz #2, you were asked to write a for loop that computes the following sum:

$$ S_n = \sum_{k=1}^n k^2 = 1^2 + 2^2 + 3^2 + 4^2 + \cdots + n^2$$

Complete the code below to define a **function** that computes the value of $S_n$ for any positive integer $n$. **Your function must compute the sum using a for loop**. Then call your function with the inputs $n = 9$, $n = 13$, and $n = 19$, printing the value of each output computed by your function to the screen.

**Hint:** You can verify your function is working by comparing the output with the algebraic expression:

$$ S_n = \sum_{k=1}^n k^2 = \frac{n(n+1)(2n+1)}{6}$$


````python
# complete the function and then call it with 3 different inputs 
def sum_of_squares(n):
    S = 0
    for k in range(1, n + 1):
        S = S + k**2
    return S

print(sum_of_squares(9))   
print(sum_of_squares(13))  
print(sum_of_squares(19)) 

````

- The second argument `col_name` is a string that corresponds to a column/variable name of the dataset. For the files in the data folder, this could be the strings "Energy_consumption", "CO2_emission", or "GDP" (computing the average of the years variable doesn't really make sense in this case).

Complete the function below so that it:

1. Reads in the .csv file corresponding to `file_path` using Pandas, converting it to a DataFrame.

2. Computes the average of the column corresponding to the string `col_name`

3. Returns the value of the average.


**Note**: From the Quiz #1 solutions, we know that we should avoid using `np.average` in our function.

````python
import pandas as pd

# Complete this function
def compute_avg(file_path, col_name):
     df = pd.read_csv(file_path)
     col = df[col_name]
     total = 0 
     count = 0 

     for val in col: 
          if pd.notna(val):
               total += val
               count += 1
     if count > 0:
        return total/count
     else: 
         return None


````
