#### **5)**. With the random library, you can simulate a dice roll by randomly generating an integer between 1 and 6, using the line of code 

 ``` python
    roll = random.choice([1,2,3,4,5,6])
```
#### (you can choose a different variable name than "roll" of course).

#### Using a `while` loop, simulate an experiment where you see how many dice rolls it takes to roll a 6. After the loop is finished, the code should print how many rolls it took.

````python
is_six = False

# SOLUTION 

import random 
roll_values = [1, 2, 3, 4, 5, 6] 

num_rolls = 0 

while not is_six:
    roll = random.choice(roll_values)
    num_rolls = num_rolls + 1  # we just rolled the die, should increase num_rolls
    if roll == 6:
        is_six = True 

print("Number of Rolls to hit a six:", num_rolls)
````

````python
# alternate approach

keep_rolling = True 

import random 
roll_values = [1, 2, 3, 4, 5, 6] 

num_rolls = 0 

while keep_rolling:
    roll = random.choice(roll_values)
    num_rolls = num_rolls + 1  # we just rolled the die, should increase num_rolls
    if roll == 6:
        keep_rolling = False 

print("Number of Rolls to hit a six:", num_rolls)
````

#### **6)**. Below, we create a random `numpy` array with positive and negative numbers. Using a `for` loop with a `continue` statement, compute the sum of all the **positive** elements of the array

````python
import numpy as np
N = 25
random_array = np.random.uniform(low = -5, high = 10, size = N)

# SOLUTION

my_sum = 0 
for idx in range(N):
    next_number = random_array[idx]
    if next_number <= 0:
        continue 
    my_sum = my_sum + next_number 

print("my_sum:", my_sum)
print("expected solution:", np.sum(random_array[random_array > 0]))
````

#### **7)**. In 1734, Leonhard Euler discovered the following identity:

$$ \frac{\pi^2}{6} = \frac{1}{1^2} + \frac{1}{2^2} + \frac{1}{3^2} + \frac{1}{4^2} + \cdots $$

#### The sum on the right is an **infinite** series. Of course we cannot perform infinitely many tasks on a computer.

#### What we **can** do is terminate the summation once the fraction $\dfrac{1}{k^2}$ becomes smaller than a pre-specified tolerance. Compute a truncated approximation to $\dfrac{\pi^2}{6}$ using a `for` loop with a `break` statement

````python
tol = 1e-6 # currently set to 10^(-6). Choose other values

# SOLUTION

approximation = 0

max_terms = 1_000_000_000  # 1 billion = 1,000,000,000

for k in range(1, max_terms + 1):
    new_term = 1 / k**2
    approximation = approximation + new_term 
    if new_term < tol:
        break 

print("Approximation to pi squared divided by six:")
print(approximation)

#### BONUS: Use approximation to pi^2 / 6 to compute approximation to pi 

pi_approx = np.sqrt( 6 * approximation )

print()
print("BONUS! Approximation to pi:")
print(pi_approx)
````
