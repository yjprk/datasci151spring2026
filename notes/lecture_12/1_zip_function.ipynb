How could we use a loop to print the following string message?

`"The movie [name of movie] has an IMDB rating of [rating of movie]"`

with the title and rating substituted in the appropriate place?

Here is one possible approach:

````python
movie_titles = ["The Umbrellas of Cherbourg", "M3gan", "Psycho"]
movie_ratings = [7.8, 6.3, 8.5]

for index in range(len(movie_titles)):
    title = movie_titles[index]
    rating = movie_ratings[index]
    print("The movie " + title + " has an IMDB rating of " + str(rating))
````

<font size = "4">

We could have also looped over one of the lists directly, and manually updated our own `index` variable:

````python
movie_titles = ["The Umbrellas of Cherbourg", "M3gan", "Psycho"]
movie_ratings = [7.8, 6.3, 8.5]

print("Looping over movie_titles:")
index = 0
for title in movie_titles:
    rating = movie_ratings[index]
    print("The movie " + title + " has an IMDB rating of " + str(rating))
    index = index + 1

print()

print("Looping over movie_ratings:")
index = 0
for rating in movie_ratings:
    title = movie_titles[index]
    print("The movie " + title + " has an IMDB rating of " + str(rating))
    index = index + 1
````

<font size = "4">

This manual `index` variable is kind of awkward. We have to plan out what value it should start with and when it should be updated in the loop.

It would be nice if we could somehow combine the first two for loops we saw:

```python
    for title in movie_titles:
        print(title) 

    for rating in movie_ratings:
        print(rating)
```

and have **two** loop variables representing the title and the rating, without introducing an extra index variable.

Fortunately, we can do this with the `zip` command:

````python
movie_titles = ["The Umbrellas of Cherbourg", "M3gan", "Psycho"]
movie_ratings = [7.8, 6.3, 8.5]

for title, rating in zip(movie_titles, movie_ratings):
    print("The movie " + title + " has an IMDB rating of " + str(rating))
````

<font size = "4">

We can swap the order of the two lists inside the `zip` command, as long as we keep our loop variables consistent:

````python
movie_titles = ["The Umbrellas of Cherbourg", "M3gan", "Psycho"]
movie_ratings = [7.8, 6.3, 8.5]

for rating, title in zip(movie_ratings, movie_titles):
    print("The movie " + title + " has an IMDB rating of " + str(rating))
````

<font size = "4">

You can use `zip` on more than two lists:

````python
movie_titles = ["The Umbrellas of Cherbourg", "M3gan", "Psycho"]
movie_ratings = [7.8, 6.3, 8.5]
movie_years = ["1964", "2022", "1960"]

for title, rating, year in zip(movie_titles, movie_ratings, movie_years):
    message = "The movie " + title + " (" + year + ") has an IMDB rating of " + str(rating)
    print(message)
````

<font size = "4">

Here's an example where we `zip` over 4 lists:

````python
movie_titles = ["The Umbrellas of Cherbourg", "M3gan", "Psycho"]
movie_ratings = [7.8, 6.3, 8.5]
movie_years = ["1964", "2022", "1960"]
movie_directors = ["Jacques Demy", "Gerard Johnstone", "Alfred Hitchcock"]

for title, rating, year, director in zip(movie_titles, movie_ratings, movie_years, movie_directors):
    message = "'" + title + "' (" + year + ") was directed by " + director + ". IMDB rating: " + str(rating)
    print(message)
````

<font size = "4">

You can use `zip` with combinations of other sequence objects like NumPy arrays and tuples:

````python
title_list = ["The Umbrellas of Cherbourg", "M3gan", "Psycho"]

import numpy as np 
rating_array = np.array([7.8, 6.3, 8.5])

year_tuple = ("1964", "2022", "1960")

director_list = ["Jacques Demy", "Gerard Johnstone", "Alfred Hitchcock"]

for title, rating, year, director in zip(title_list, rating_array, year_tuple, director_list):
    message = "'" + title + "' (" + year + ") was directed by " + director + ". IMDB rating: " + str(rating)
    print(message)
````
