"Recoding" variables means changing/re-expressing existing variables into a new set of values that may be more useful for data analysis.

Examples:

- A variable called "age" consists of integers beginning at 0 (how old the person is in terms of years). We can transform this into a categorical variable with the values "child", "teenager", "young adult", "adult", "senior"

- A variable called "submission_time" records the time that a student submitted an assignment to Gradescope. We can transform this into a categorical variable with the values "on time" and "late"

- A variable called "candidate_preference" is taken from a survey of people asking which of two presidential candidates they will vote for. This is a categorical variable with the values "Candidate A", "Candidate B", "Don't Know", "Refused", "Blank", "Other Candidate". This is recoded to another categorical variable with only 3 values "Candidate A", "Candidate B", and "Neither".

- We convert a variable "height" measured in inches, to a corresponding variable measuring height in centimeters.

- A variable called "lat" consists of floating point numbers consisting of a geographical coordinate measuring the distance north or south of the Equator. We change this into a categorical variable with the values "north" or "south".

<font size = "4">

We'll perform this last one in **3 different ways** using the circuits data we cleaned last time. In each way, we will consider as "south" those circuits with $-90 < \textrm{lat} \leq 0$ and "north" as circuits with $0 < \textrm{lat} \leq 90$

```python
import pandas as pd

circuits = pd.read_csv("f1_clean_data/circuits_final.csv")

circuits.head()
```

<font size = "4">

Here we do it manually with a for loop and an if/else statement:

```python
nrows, ncols = circuits.shape 

col_no = 5 # column 5 is "lat"
hemisphere = []

for k in range(nrows):
    lat_val = circuits.iloc[k, col_no]
    if lat_val <= 0:
        hemisphere.append("South")
    else:
        hemisphere.append("North")

# could overwrite "lat", but is almost always better to keep the original variable
circuits["hemisphere"] = hemisphere

# Note: I made sure the data in this column didn't need to be cleaned first!!
circuits.head()
```

<font size = "4">

Let's drop the "hemisphere" column to verify that the next approach we try actually works

```python
circuits.drop(columns = "hemisphere",inplace = True)
circuits.head()
```

<font size = "4">

In the next cell, I *attempt* to use `.apply` to recode the data. 

**What error did I make and how should I fix it?**

```python
def check_hemisphere(lat):
    if lat <= 0: # this line used to say "lat_val <= 0"
        return "South" 
    else:
        return "North"

circuits["hemisphere"] = circuits["lat"].apply(check_hemisphere)

circuits.head()
```

```python
# dropping the "hemisphere" column again:
circuits.drop(columns = "hemisphere",inplace = True)
circuits.head()
```

<font size = "4">

Finally, I use the function `pd.cut` which is used to segment data into "intervals" or "bins".

Remember, we are recoding the latitute variable based on the following criteria:

- If $-90 < \textrm{lat} \leq 0$, the circuit is labeled "south"

- If $0 < \textrm{lat} \leq 90$, the circuit is labeled "north"

```python
lat_bins = [-90, 0, 90]

lat_labels = ["South", "North"]
circuits["hemisphere"] = pd.cut(circuits["lat"], 
                                bins = lat_bins,
                                right = True,
                                labels = lat_labels)

# The "right" command indicates that the right endpoint of each interval
# is *inclusive*, i.e. "less than or equal to"


circuits.head()
                            
```

<font size = "4">

Alternatively, we can implement it more robustly by recoding the latitude variable by:

- If $-\infty < \textrm{lat} \leq 0$, the circuit is labeled "south"

- If $0 < \textrm{lat} \leq \infty$, the circuit is labeled "north"

(Important in the unlikely event an F1 racetrack is ever built directly at the South Pole)

```python
import numpy as np

lat_bins = [-np.inf, 0, np.inf]

lat_labels = ["South", "North"]
circuits["hemisphere"] = pd.cut(circuits["lat"], 
                                bins = lat_bins,
                                right = True,
                                labels = lat_labels)


# Note: Since we set lat_bins = [-np.inf, 0, np.inf]
# then intervals are "Less than or equal to 0" and "Above 0"

circuits.head()
                      
```
