```python
display(carfeatures["cylinders"])
```

```python
#cross tabulation = contingency table 
table = pd.crosstab(index = carfeatures["cylinders"],columns = ["count"])
display(table)
```

```python
table = pd.crosstab(carfeatures["cylinders"], "count")
# frequency table
display (table)
```

```python
table_2 = pd.crosstab(index = carfeatures ["cylinders"], columns = carfeatures["mpg"]) 
# this is a contingency table which is why you need to add carfeatures in front of mpg 
display(table_2)
```

```python
plt.scatter(x = carfeatures['weight'], y = carfeatures['mpg'])
plt.xlabel("Weight")
plt.ylabel("MPG")
plt.title("Relationship between weight \n and gas mileage of automboiles") #the /n continues the title below 
plt.show()
#if plt.show isnt written above, then the two scatterplots will be shown on one graph. if you're going to make 2 different plots make sure to write plot.show twice 

plt.scatter(x = carfeatures['acceleration'], y = carfeatures['mpg'])
plt.xlabel("Acceleration")
plt.ylabel("MPG")
plt.title("Relationship between acceleration \n and gas mileage of automboiles") #the /n continues the title below 
plt.show()

```

```python
plt.scatter(x = carfeatures['acceleration'], y = carfeatures['mpg'])
plt.xlabel("Acceleration")
plt.ylabel("MPG")
plt.title("Relationship between acceleration \n and gas mileage of automboiles") #the /n continues the title below 
plt.show()
```
