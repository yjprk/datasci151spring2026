Below, I read in the dataset "features.csv", which contains data for different make/models of cars. After reading it in, we generate a table of basic descriptive statistics (rounded to two decimal digits)

```python
import pandas as pd

df_cars = pd.read_csv("data/features.csv")

summary = df_cars.describe().round(2)

display(summary)
```

- Next, we generate a scatterplot of car weight vs. car mileage.

- It is a good idea to motivate why you are doing this. For example, you might briefly explain that it is a reasonable hypothesis that heavier cars have worse gas mileage.

```python
import matplotlib.pyplot as plt

plt.scatter(x = df_cars['weight'], y = df_cars['mpg'])
plt.xlabel("Weight")
plt.ylabel("MPG")
plt.title("Relationship between weight \n and gas mileage of automobiles")
plt.show()
```
