## <span style="color:red; font-size:25px"> I. Boolean objects </span>

<font size = "4">

Out of the many built-in objects in Python, two of them are of "Boolean" type. These are the objects ``True`` and ``False``

These should not be confused with the strings "True" and "False"

````python
# Boolean Objects
print(type(True))
print(type(False))

print()

# Adding quotes results in string objects
print(type("True"))
print(type('False'))
````

<font size = "4">

What is a Boolean object? It is just an object that can take one of two possible values. In this case, `True` and `False`.

We can assign these values to variables directly:

````python
x = True 

my_statement = False 
````

<font size = "4">

But more commonly, we obtain Boolean values in other, indirect ways.

## <span style="color:red; font-size:25px"> II. Comparison Operators </span>

<font size = "4">


- One of the most common ways Boolean objects show up is in the use of **comparison operators**.
- These are operators used to compare two values, and evaluate to either `True` or `False`.
- There are six basic comparison operators. The first four are:

    - ``>`` - Greater than

    - ``<`` - Less than
    - ``>=`` - Greater than or equal to
    - ``<=`` - Less than or equal to

````python
# Test if one number is greater than the other
print(5 > 3)

print(3 > 9)

print(4 + 3 > 6)
````

````python
# Test if one number is less than the other
print(0 < -2.5)

print(1 < 1.01)

print(30/10 < 1)
````

````python
# Test if one number is greater that or equal to the other
print(10 >= 1)

print(8 >= 50)

print(2**3 >= 10)
````

````python
# Greater than OR equal to!
print(5 > 5) # this is false
print(5 >= 5) # this is true, since 5 equals 5
````

````python
# Test if one number is less than or equal to the other
print(10 <= 5 + 5)

print(8 <= 8.1)

print(2**3 <= 2*3)
````

<font size = "4">

Two more basic comparison operators
- ``==`` - Equal

- ``!=`` - Not equal

**Warning**: Do not confuse the assignment operator ``=`` with the comparison operator that tests for equality ``==``

````python
# Does 5*3 equal 15? (Yes)
print(5*3 == 15)

# Does 3 - 2 equal 7? (No)
print(3 - 2 == 7)
````

````python
# Is 5*3 + 1 different from 20? (yes)
print(5*3 + 1 != 20)

# Is 10/5 different from 2? (no)
print(10/5 != 2)
````

<font size = "4">

In many cases, we use a comparison operator with one or more variables 

````python
a = 7
b = 3
c = 5

print(a*b > c**2)


x = np.sin(12*np.pi/7)
print(x < 0)
````

## <span style="color:red; font-size:25px"> III. Strings and Lists </span>

<font size = "4">

- The "equals to" comparison and "not equals to" comparison extend naturally to strings and lists

- If every character in a string matches another string **exactly**, then they are equal.
- If every element in a list matches **every** element in another list, then they are equal. **Order of the elements matters**


````python
x = "Hello World!"
y = "Hello World !"

print(x == "Hello World!")
print(y == "Hello World!")

print(x != y)

print('ab' == 'ba')
````

````python
list_1 = ["blue", "red", "green"]
list_2 = ["blue", "red", "green"]
list_3 = ["red", "green", "blue"]
list_4 = ["red", "green", "blue", "violet"]

print(list_1 == list_2)
print(list_1 == list_3) # False: different ordering
print(list_3 == list_4) # False: different number of elements
print(list_3 != list_4) # True
````

````python
list_1 = [1, 2, 3]
list_2 = [1, 2, 3]
list_3 = [2, 3, 1]
list_4 = [2, 3, 1, 4]

print(list_1 == list_2)
print(list_1 == list_3) # False: different ordering
print(list_3 == list_4) # False: different number of elements
print(list_3 != list_4) # True
````

<font size = "4">

The other comparison operators `<`, `>`, `<=`, and `>=` can also be used for strings and lists. However, the behavior is not obvious. See the exercises for more.

## <span style="color:red; font-size:25px"> V. The `in` keyword </span>

<font size = "4">

This keyword allows us to test whether
- A character appears in a string
- An object appears in a list
- A number appears in a numpy array

````python
# The first way to use the "in" command is to check whether a word is contained 
# in a sentence. This can be useful if you're trying to search for patterns

sentence = "The Federal Reserve makes forecasts about many economic outcomes"


keyword = "economic"
print(keyword in sentence)

keyword = "bank"
print(keyword in sentence)

search_word = "Federal" 
print(search_word in sentence)

important_word = "federal"
print(important_word in sentence)

print("F" in sentence)
````

````python
# Use the condition "in" to test whether a word is part of a list
# when the right-hand size is a list "word in list" will search for an exact match

current_month  = "February"

list_summer_months = ["June","July","August"]

print(current_month in list_summer_months)
print('June' in list_summer_months)
print("J" in list_summer_months)
````

````python
# Use the keyword "in" to see if a number belongs to a NumPy Array
# (Also works for lists)

s = np.array([1.0, 3.0, -17, 12])
print(1 in s)

t = [0, 22, "PAIS 557"]

print(2 in t)
````

## <span style="color:red; font-size:25px"> VI. Logical Operators </span>

<font size = "4">

- Logical operators allow you to combine conditional statements

- The three logical operators are `not`, `and`, `or`

````python
age  = 22

# True or false? This person can vote in the U.S.

# 1st way to test it:
print(age >= 18)
````

````python
age  = 22

# True or false? This person can vote in the U.S.

# 2nd way to test it:
print(not age < 18)
````

````python
# Sometimes introducing an intermediate variable can make things more clear

age = 22 

statement = age < 18
print(statement) 

print(not statement)
````

````python
price = 24.5

list_prices = [32.5, 13, 0.99, 15.96]

# True or false? The price is greater than 10 dollars 
# and is one of the elements of list_prices

print(price > 10 and price in list_prices)
````

````python
price = 24.5

list_prices = [32.5, 13, 0.99, 15.96]

statement_1 = price > 10 
statement_2 = price in list_prices

print(statement_1)
print(statement_2)
print(statement_1 and statement_2)
````

<font size = "4">

**Both** statements must be true for [statement_1] and [statement_2] to be true

To test if **at least** one statment is true, use the `or` operator

````python
price = 24.5

list_prices = [32.5, 13, 0.99, 15.96]

statement_1 = price > 10 
statement_2 = price in list_prices

print(statement_1)
print(statement_2)
print(statement_1 or statement_2)
````

````python
st_1 = -1 >= 10
st_2 = 12 > 13

print(st_1)
print(st_2)
print(st_1 or st_2)
````

<font size = "4">

While not necessary, can separate statements with parentheses for readibility

````python
x = 13
y = -5
z = False

P = (x > y) and (not z) and (x*y < 0)
print(P)

Q = (x == 5) or (y > 10) or (x + y > 0)
print(Q)

# try changing some of these conditions to experiment
````

**Test expression**

- The "body" of the `if` statement needs to be indented

**If the condition is true, the code in the indented block will be run.**

**If the condition is false, then the indented block is skipped.**

````python
good_movies = [] # create empty list

cut_off = 7.0 # minimum rating for "good movie"

# Define variables for 3 movie titles and their corresponding IMDB rating
title_1 = "The Umbrellas of Cherbourg"
imdb_rating_1 = 7.8 

title_2 = "M3gan" 
imdb_rating_2 = 6.3 

title_3 = "Psycho" 
imdb_rating_3 = 8.5 

# Determine which movies to add to list

if imdb_rating_1 >= cut_off:
    good_movies = good_movies + [title_1] # add movie to list by concatenation 

if imdb_rating_2 >= cut_off:
    good_movies = good_movies + [title_2] # add movie to list by concatenation 

if imdb_rating_3 >= cut_off:
    good_movies = good_movies + [title_3] # add movie to list by concatenation 


print(good_movies)
````

Now if the statement is `False`, the first code block will be skipped, and the second will be run instead.

````python
# create empty lists
good_movies = [] 
other_movies = []

cut_off = 7.0 # minimum rating for "good movie"

# Define variables for 3 movie titles and their corresponding IMDB rating
title_1 = "The Umbrellas of Cherbourg"
imdb_rating_1 = 7.8 

title_2 = "M3gan" 
imdb_rating_2 = 6.3 

title_3 = "Psycho" 
imdb_rating_3 = 8.5 


# Divide movies between the two lists

if imdb_rating_1 >= cut_off:
    good_movies = good_movies + [title_1] # add movie to list by concatenation 
else:
    other_movies = other_movies + [title_1]

if imdb_rating_2 >= cut_off:
    good_movies = good_movies + [title_2] # add movie to list by concatenation 
else:
    other_movies = other_movies + [title_2]

if imdb_rating_3 >= cut_off:
    good_movies = good_movies + [title_3] # add movie to list by concatenation 
else:
    other_movies = other_movies + [title_3]

print(good_movies)
print(other_movies)
````

````python
# create empty lists
good_movies = [] 
ok_movies = []
bad_movies = []

cut_off_1 = 7.0 # minimum rating for "good movie"
cut_off_2 = 6.0


title_1 = "The Umbrellas of Cherbourg"
imdb_rating_1 = 7.8 

title_2 = "M3gan" 
imdb_rating_2 = 6.3 

title_3 = "Psycho" 
imdb_rating_3 = 8.5 

title_4 = "Melania"
imdb_rating_4 = 1.3

# Divide movies between the three lists
# print a warning if we encounter a bad movie

if imdb_rating_1 >= cut_off_1:
    good_movies = good_movies + [title_1] # add movie to list by concatenation 
elif imdb_rating_1 >= cut_off_2:
    ok_movies = ok_movies + [title_1]
else:
    bad_movies = bad_movies + [title_1]
    print("Warning! Bad movie alert:", title_1)

if imdb_rating_2 >= cut_off_1:
    good_movies = good_movies + [title_2] # add movie to list by concatenation 
elif imdb_rating_2 >= cut_off_2:
    ok_movies = ok_movies + [title_2]
else:
    bad_movies = bad_movies + [title_2]
    print("Warning! Bad movie alert:", title_2)


if imdb_rating_3 >= cut_off_1:
    good_movies = good_movies + [title_3] # add movie to list by concatenation 
elif imdb_rating_3 >= cut_off_2:
    ok_movies = ok_movies + [title_3]
else:
    bad_movies = bad_movies + [title_3]
    print("Warning! Bad movie alert:", title_3)


if imdb_rating_4 >= cut_off_1:
    good_movies = good_movies + [title_4] # add movie to list by concatenation 
elif imdb_rating_4 >= cut_off_2:
    ok_movies = ok_movies + [title_4]
else:
    bad_movies = bad_movies + [title_4]
    print("Warning! Bad movie alert:", title_4)


print(good_movies)
print(ok_movies)
print(bad_movies)

````
